```python
from matplotlib import style
style.use("ggplot")
x=[7,5,6]
y=[6,7,8]
x2=[3,4,5]
y2=[1,6,4]
plt.plot(x,y,'g', label='LineOne' ,linewidth=5)
plt.plot(x2,y2,'b', label='LineTwo' ,linewidth=5)
plt.title("Info")
plt.ylabel("x axis")
plt.xlabel("y axis")

plt.grid(True,color="b")
plt.show()
```

```python
from matplotlib import pyplot as plt
plt.bar([1,1,3,4],[5,6,7,8], label="Example one")
plt.bar([5,3,2,1],[1,7,5,3], label="Exmaple Two", color="g")
plt.legend()
plt.title("info")
plt.ylabel("x axis")
plt.xlabel("y axis")
plt.legend()
plt.show()
```

```python
from matplotlib import pyplot as plt
population_ages = [22,33,55,66,77,88,24,64,24,87,32,87]
bins=[45,65,86,24,75,25,97,32]
plt.hist(population_ages,bins,histtype='bar', rwidth=0.8)
plt.legend()
plt.title("info")
plt.ylabel("x axis")
plt.xlabel("y axis")
plt.show()
```
