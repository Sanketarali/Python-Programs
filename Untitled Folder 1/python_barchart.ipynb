```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company=['tcs', 'infosys', 'wipro']
revenue=[50,80,46]
plt.title('revenue')
plt.xlabel('company')
plt.ylabel('revenue')
plt.bar(company, revenue , label='revenue', color='red')
plt.legend()
```

```python
company=['tcs', 'infosys', 'wipro']
revenue=[50,80,46]
plt.title('revenue')
plt.xlabel('company')
plt.ylabel('revenue')
plt.barh(company, revenue , label='revenue', color='red')
plt.legend()
```
